```python
import geopandas as gpd
import matplotlib.pyplot as plt
import folium

# Coordenadas das cidades (latitude, longitude)
cities = {
    "Suape-PE": (-8.3975, -34.9633),
    "MaceiÃ³-AL": (-9.6658, -35.7353),
    "MaracanaÃº-CE": (-3.8766, -38.6256),
    "MacaÃ­ba-RN": (-5.8522, -35.3552),
    "SÃ£o LuÃ­s-MA": (-2.5391, -44.2829),
    "SimÃµes Filho-BA": (-12.7844, -38.3906),
    "VitÃ³ria da Conquista-BA": (-14.8615, -40.8442),
    "VÃ¡rzea Grande-MT": (-15.6467, -56.1325),
    "BelÃ©m-PA": (-1.4558, -48.5044),
    "Benevides-PA": (-1.3617, -48.2434),
    "Manaus-AM": (-3.1190, -60.0217),
    "Porto Velho-RO": (-8.7608, -63.9004),
}

# Criar o mapa do Google Maps com Folium
my_map = folium.Map(location=[-10.0, -55.0], zoom_start=4)

# Adicionar marcadores para as cidades
for city, (lat, lon) in cities.items():
    folium.Marker(
        location=[lat, lon],
        popup=city,  # Adiciona um pop-up com o nome da cidade
        icon=folium.Icon(color="red", icon="fa-location-arrow", prefix="fa")
    ).add_to(my_map)

# Salvar o mapa como um arquivo HTML
my_map.save("/content/drive/MyDrive/MOAB/Python Colab\Mapa Grupo SOLAR COCA COLA.html")

# Visualizar o mapa no navegador (opcional)
my_map
```
